```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7,8,9]
max_t = [10,20,30,40,50,60,70,80,90]
AVG_t = [45,12,65,63,75,32,62,25,89]
min_t = [5,10,15,20,25,30,35,40,45]
plt.plot(days,max_t,label="Max Temperature")
plt.plot(days,AVG_t,label="Avg Temperature")
plt.plot(days,min_t,label="Min Temperature")
plt.legend(loc= "best")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather Graph")
```

```python
import matplotlib.pyplot as plt
Football_Player = ["Casillas","Cristaino","Messi","Neymar","Mbappe","Ramos"]
salary = [120,450,435,222,302,180]
plt.bar(Football_Player,salary,label="Salary",width=0.4)
plt.title("Top Richest FootBall Players")
plt.xlabel("Players Name -------------------->")
plt.ylabel("Player's Salary  -------------------->")
plt.legend()
```

```python
import matplotlib.pyplot as plt
SugarLevel = [12,30,45,80,43,75,62,22,30,42,18,40,42,23,65,44,23,54,32,34]
BodyTemperature = [10,20,30,40,50,60,70,80,90]
# plt.hist(SugarLevel)
plt.hist([SugarLevel,BodyTemperature],bins=[25,50,75,100],color=["g","b"],label=['men','women'],orientation='horizontal')
plt.xlabel('SugarLevel ------------->')
plt.ylabel('BodyTemperature---------->')
plt.title('Health Chart')
plt.legend()
```

```python
import matplotlib.pyplot as plt
SugarLevel = [12,30,45,80,43,75,43,64]
names=["ua","dk","sr","ks","s","kar","ms","rp"]
plt.pie(SugarLevel,labels=names,autopct='%0.0f%%',shadow=True,startangle=45)
plt.axis('equal')
plt.show()
```
